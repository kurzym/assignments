```
# liczba ludności w miastach vs śmiertelność

import plotly.express as px

fig = px.bar(Wskaźniki_mnpp, x='JST', y='Liczba Ludności', color='Zgony na 1000 mieszkańców', 
             labels={'Miasta': 'Miasta ich ludność i śmierć'}, height=650, hover_name='JST', 
             title = "liczba ludności w miastach vs śmiertelność")
fig.show()

```

```
# interaktywny wykres ukazujący kto z nas był, w którym mieście
fig = px.scatter(Wskaźniki_mnpp, x='JST', y='czy Magda była?', size='Czy Tomek był?', color='czy Rafał był?', size_max=15, hover_name='JST', range_y= (-0.5,1.5), title = "Kto z nas w jakim mieście był?")
fig.show() 
```

```
# plot wskaźnik obciążenia demograficznego vs starość w zależności od liczby uczniów na 1 etat nauczyiela
fig = px.scatter(Wskaźniki_mnpp, x='Wskaźnik obciążenia demograficznego (przedprodukcyjny/produkcyjny)', y='Wskaźnik starości (emeryci/ogółem)', color='Liczba uczniów na 1 etat nauczycieli', size = 'Liczba Ludności', size_max=29, hover_name='JST', range_y= (0.18, 0.33), title = "Wskaźnik obciążenia demograficznego vs starość w zależności od liczby uczniów na 1 etat nauczyciela")
fig.show() 
```
